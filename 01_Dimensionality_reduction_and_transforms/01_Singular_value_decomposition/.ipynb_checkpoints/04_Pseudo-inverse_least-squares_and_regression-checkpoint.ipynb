```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Figure 1.9
x = 3 # True slope
a = np.arange(-2,2,0.25)
a = a.reshape(-1, 1) # Turns a into a two-dimensional array
b = x*a + np.random.randn(*a.shape) # Add noise

u, s, vh = np.linalg.svd(a, full_matrices=False)
xtilde = vh.T @ np.linalg.inv(np.diag(s)) @ u.T @ b # Least-square fit

plt.plot(a, x*a, color='k') # True relationship
plt.plot(a, b, 'x', color='r') # Noisy measurements
plt.plot(a, xtilde*a, '--', color='b') # Plot fit
plt.show()
```

```python
# Figure 1.10
# Load dataset
A = np.loadtxt('hald_ingredients.csv',delimiter=',')
b = np.loadtxt('hald_heat.csv',delimiter=',')

# Solve Ax=b using SVD
u, s, vh = np.linalg.svd(A,full_matrices=0)
x = vh.T @ np.linalg.inv(np.diag(s)) @ u.T @ b

plt.plot(b, color='k') # Plot data
plt.plot(A@x, '-o', color='r') # Plot fit
plt.show()
```

```python
# Figure 1.11
# Load dataset
H = np.loadtxt('housing.data')
b = H[:,-1] # housing values in $1000s
A = H[:,:-1] # other factors

# Pad with ones for nonzero offset
A = np.pad(A,[(0,0),(0,1)],mode='constant',constant_values=1)

# Solve Ax=b using SVD
U, S, VT = np.linalg.svd(A,full_matrices=0)
x = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b

plt.plot(b, color='k') # Plot data
plt.plot(A@x, '-o', color='r') # Plot fit
plt.show()

sort_ind = np.argsort(H[:,-1])
b = b[sort_ind] # sorted values
plt.plot(b, color='k') # Plot data
plt.plot(A[sort_ind,:]@x, '-o', color='r') # Plot fit
plt.show()
```

```python
# Figure 1.12
A_mean = np.mean(A,axis=0)
A_mean = A_mean.reshape(-1, 1)

A2 = A - np.ones((A.shape[0],1)) @ A_mean.T

for j in range(A.shape[1]-1):
    A2std = np.std(A2[:,j])
    A2[:,j] = A2[:,j]/A2std
    
A2[:,-1] = np.ones(A.shape[0])

u, s, vh = np.linalg.svd(A2,full_matrices=0)
x = vh.T @ np.linalg.inv(np.diag(s)) @ u.T @ b
x_tick = range(len(x)-1)+np.ones(len(x)-1)
plt.bar(x_tick,x[:-1])
plt.xticks(x_tick)
plt.show()
```
