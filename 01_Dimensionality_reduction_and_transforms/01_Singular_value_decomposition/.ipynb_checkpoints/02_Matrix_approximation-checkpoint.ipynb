```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

# Original image
A = mpimg.imread('dog.jpg')
plt.imshow(A)
plt.axis('off')
plt.show()
```

```python
# Figure 1.3
X = np.mean(A, -1) # RGB -> grayscale

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()

u, s, vh = np.linalg.svd(X, full_matrices=False)
s = np.diag(s) # sparse, diagonal matrix

for r in (5,20,100):
    Xapprox = u[:,:r] @ s[0:r,:r] @ vh[:r,:]
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.show()
```

```python
# Figure 1.4
plt.semilogy(np.diag(s))
plt.show()

plt.plot(np.cumsum(np.diag(s))/np.sum(np.diag(s)))
plt.show()
```

```python
# Figure 1.5
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()

# These plots look backwards from the textbook
img = plt.imshow(X @ np.asmatrix(X).H)
img.set_cmap('gray')
plt.axis('off')
plt.show()

img = plt.imshow(np.asmatrix(X).H @ X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```
